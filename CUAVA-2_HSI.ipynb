```python
import numpy as np

# Load the CUAVA-2 file (the data is stored in files with extension *.npz as received from Xueliang - i think it is compressed)
b = np.load('C:/Users/al-na/Desktop/Kevin/HSI data/image_20230920_0652108054.npz', allow_pickle=True) 

# what is in the file
print(b.files)
```

```python
#Inspecting the arrays in file
array1 = b['arr_0'] # this array has the HSI data
array2 = b['arr_1'] # this array has all the meta data like exposure time, number of frames etc

# Print them
print(array1)
print(array2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Array1 has the spatial data 
array1 = b['arr_0']

# plot the spatial image
fig, ax = plt.subplots(figsize=(16, 8)) 
im = ax.imshow(array1, cmap='viridis')  # array1 has the spatial data
ax.set_title('CUAVA-2 HSI Image', fontweight='bold', fontsize=16)  

#colorbar
cbar = fig.colorbar(im, ax=ax, orientation='vertical')
cbar.set_label('Intensity')
plt.show()
```

```python
# Select spatial coordinate for extraction
y = 850   # Example y-coordinate

# Extract spectrum at the selected spatial location (y-coordinate)
spectrum = array1[y, :]

# Plot spectrum with dashed line
plt.figure(figsize=(10, 6))
plt.plot(spectrum, linestyle='-')  # Set linestyle to '--' for dashed line
plt.title('Spectrum at Spatial Pixel y = {}'.format(y))
plt.xlabel('Spectral Index')
plt.ylabel('Intensity')
plt.grid()
plt.show()
```
